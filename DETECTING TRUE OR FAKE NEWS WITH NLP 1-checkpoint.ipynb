## UPLOADING DATASET

```python
import pandas as pd
fake_news=pd.read_csv("C:\\Users\\HP\\Desktop\\Fake.csv")
fake_news.head()
```

```python
real_news=pd.read_csv("C:\\Users\\HP\\Desktop\\True.csv")
real_news.head()
```

```python
#Adding a Label colunm to indicate whether each article is fake or not 
fake_news['label'] = 'FAKE'
real_news['label'] = 'REAL'

#Concatenate the two Dataframe into one
df = pd.concat([fake_news, real_news], axis=0)

#Reset the index of dataframe
df = df.reset_index(drop=True)
#Display the first 5 rows of the new Dataframe
df.head()
```

## CLEANING DATASET

```python
#Checking for missing values 
df.isnull().sum()
```

```python
print(df.shape)
```

```python
df.describe()
```

```python
#Combining the text from the 'title' and 'text' columns into a new column called 'text_combined
df['text_combined'] = df['title'] + ' ' + df['text']
df.head()
```

```python
# CONVERT ALL TEXT TO LOWERCASE
df['text_combined']= df['text_combined'].str.lower()
```

```python
#Remove stop words and punctuation from the text
stop_words = stopwords.words('english') + list(string.punctuation)
df['text_combined'] = df['text_combined'].apply(lambda x: ' '.join([word for word in word_tokenize(x) if word not in stop_words]))
```

```python
#REVOVING PUNCTUATION FROM THE TEXT
import re

# Define regex pattern to match symbols and punctuation
pattern = r'[^\w\s]'

# Apply regex pattern to remove symbols and punctuation from text_combined column
df['text_combined'] = df['text_combined'].apply(lambda x: re.sub(pattern, '', x))
```

```python
#Apply lemmatization or stemming 
# Example of lemmatization using WordNetLemmatizer

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
df['text_combined'] = df['text_combined'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in word_tokenize(x)]))
```

```python
#Tokenize text and title into individual word
df['tokens'] = df['text_combined'].apply(word_tokenize)
```

## FEATURE ENGINEERING


TOPIC MODELLING

```python
import gensim
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel
from gensim.corpora import Dictionary

# Convert each document to a list of tokens using simple_preprocess
documents = [simple_preprocess(doc) for doc in df[df['label'] == 'FAKE']['text_combined']]

# Create a dictionary from the documents
dictionary = Dictionary(documents)

# Convert the documents to a bag-of-words corpus
corpus = [dictionary.doc2bow(doc) for doc in documents]

# Build an LDA model with 10 topics
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=dictionary,
                                           num_topics=10,
                                           random_state=42,
                                           update_every=1,
                                           passes=10,
                                           alpha='auto',
                                           per_word_topics=True)

# Print the top words for each topic
topics = lda_model.show_topics(num_topics=10, num_words=10, formatted=False)
for topic in topics:
    print(f"Topic {topic[0]}: {' '.join([word[0] for word in topic[1]])}")
```

Using Sentimemtal Analysing to determine the sentiment or emotion expressed in the FAKE NEWS text_combined

```python
from textblob import TextBlob

# create a function to get the sentiment of each row
def get_sentiment(row):
    text = row['text_combined']
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    return sentiment

# apply the function to the dataframe
df['sentiment'] = df.apply(get_sentiment, axis=1)

# print the head of the dataframe with the sentiment column
print(df[['text_combined', 'label', 'sentiment']].head())
```

## Training and Testing the dataset

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import SVC
from sklearn.metrics import classification_report

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['text_combined'], df['label'], test_size=0.2, random_state=42)
```

```python
# Convert the text data into numerical features using CountVectorizer
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
# Train an SVM model on the training set
svm = SVC(kernel='linear', C=1.0)
svm.fit(X_train, y_train)
```

```python
# Test the performance of the trained model on the testing set
y_pred = svm.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
#Performing a grid search 
from sklearn.model_selection import GridSearchCV
```

```python
# Define the parameter grid for the SVM model
param_grid = {'kernel': ['linear', 'rbf'], 'C': [0.1, 1, 10]}
```

```python
# Perform grid search with 5-fold cross-validation
grid_search = GridSearchCV(SVC(), param_grid, cv=5)
```

```python
grid_search.fit(X_train, y_train)
```

```python
# Print the best parameters found by grid search
print('Best parameters:', grid_search.best_params_)
```

```python
# Use the best model to make predictions on the testing set
best_svm = grid_search.best_estimator_
y_pred = best_svm.predict(X_test)
print(classification_report(y_test, y_pred))
```

The model achieved an accuracy of 1.00 (100%), which indicates that it is able to accurately classify both the fake and real news articles.

The precision, recall, and F1-score metrics are also 1.00 (100%), which means that the model is able to accurately identify both the true positives and true negatives without any false positives or false negatives.

Overall, the results suggest that the SVM model is a good choice for classifying fake and real news articles using NLP.

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
from string import punctuation
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from joblib import load

# Load the original dataset
data = pd.read_csv("C:\\Users\\HP\\Desktop\\news.csv")
```

```python
# Check for missing values
data.isnull().sum()
```

```python
# Combine the text from the 'title' and 'text' columns into a new column called 'text_combined'
data['text_combined'] = data['title'] + ' ' + data['text']
```

```python
# Save the new dataset with the combined text
data.to_csv('data_combined.csv', index=False)
```

```python
# Define a function to clean the text data
def clean_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove punctuation and stopwords, and lemmatize
    stop_words = set(stopwords.words('english')) | set(punctuation)
    lemmatizer = WordNetLemmatizer()
    cleaned_text = ' '.join([lemmatizer.lemmatize(word) for word in word_tokenize(text) if word not in stop_words])

    return cleaned_text
```

```python
# Apply the clean_text function to the text data in the original dataset
data['clean_text'] = data['text_combined'].apply(clean_text)
```
